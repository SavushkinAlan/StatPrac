#### Ex. 1
Написать функцию, которая по заданным коэффициэнтам квадратного 
<br>
трехчлена выдает его корни, если они вещественны или фразу "вещественных корней нет"

```python
def real_eq_solution(a, b, c):
    if a == 0:
        return -c/b if b else 'not solutions'
    else:
        d = b**2 - 4 * a * c
        if d >= 0:
            return ((-b + d**0.5) / (2 * a), (-b - d**0.5) / (2 * a)) if d != 0 else -b  / (2 * a)
        else:
            return 'no real roots'
```

```python
real_eq_solution(0,0,8)
```

```python
real_eq_solution(0, 7, -9)
```

```python
real_eq_solution(0, 0.3, 1)
```

```python
real_eq_solution(1, 1, 1)
```

```python
real_eq_solution(1, 2, 1)
```

#### Ex. 2
База flights содержит информацию о количестве перевезенных пассажиров в
<br>
различных месяцах. Подсчитать количество перевезенных в различные месяцы по всем
<br>
годам в сумме. Построить диаграммы рассеяния и сравнить популярность различных
<br>
месяцев среди пассажиров.

```python
df = sns.load_dataset('flights')
```

```python
df.groupby('month')[['passengers']].sum().plot(kind='bar', rot=45)
```

```python
_ = plt.figure(figsize=(16, 6))
_ = sns.scatterplot('year', 'passengers', hue='month', data=df)
```

По первому графику видно, что больше всего пассажиров летает в летние месяцы.
В осенние и зимние месяцы кол-во пассажиров было меньше
<br>
По второму графику видно, во-первых с каждым годом кол-во пассажиров в целом растет,
<br>
А во-вторых, более наглядно видно, что более популярными для перелетов являются летние меясцы
<br>
по сравнению с осатльными временами года 

```python
_ = plt.figure(figsize=(12, 6))
_ = sns.boxplot(x='month', y='passengers', data=df)
```
